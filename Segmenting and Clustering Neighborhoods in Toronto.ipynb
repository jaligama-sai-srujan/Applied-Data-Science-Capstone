### Importing Libraries

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
print("Imported!")
```

```python
extracting_data = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
wiki_data = BeautifulSoup(extracting_data, 'lxml')
wiki_data
```

### Converting Beautiful Soup to DataFrame

```python
column_names = ['Postalcode','Borough','Neighborhood']
toronto = pd.DataFrame(columns = column_names)

content = wiki_data.find('div', class_='mw-parser-output')
table = content.table.tbody
postal = 0
borough = 0
neighbor = 0

for tr in table.find_all('tr'):
    i = 0
    for td in tr.find_all('td'):
        if i == 0:
            postal = td.text
            i = i + 1
        elif i == 1:
            borough = td.text
            i = i + 1
        elif i == 2: 
            neighbor = td.text.strip('\n').replace(']','')
    toronto = toronto.append({'Postalcode': postal,'Borough': borough,'Neighborhood': neighbor},ignore_index=True)
# toronto.columns = ((toronto.columns.str).replace("\n",""))

for i in range(len(toronto)):
    for j in range(len(toronto.iloc[i])):
        toronto.iloc[i,j] = str(toronto.iloc[i,j]).replace('\n','')
toronto
```

### Data Cleaning

```python
toronto = toronto[toronto.Borough!='Not assigned']
toronto = toronto[toronto.Borough!= 0]
toronto.reset_index(drop = True, inplace = True)
i = 0
for i in range(0,toronto.shape[0]):
    if toronto.iloc[i][2] == 'Not assigned':
        toronto.iloc[i][2] = toronto.iloc[i][1]
        i = i+1
toronto
```

```python
### Grouping data as More than one neighborhood can exist in one postal code area, Then droupping nan's

df = toronto.groupby(['Postalcode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
df.dropna(inplace=True)
df
```

```python
df.describe()
```

```python

def neighbors(grouped):    
    return ', '.join(sorted(grouped['Neighborhood'].tolist()))
                    
group = df.groupby(['Postalcode', 'Borough'])
new_df = group.apply(neighbors).reset_index(name='Neighborhood')
new_df
```

```python
### printing number of rows in the frame
print(new_df.shape)
```

```python
new_df.to_csv('toronto.csv',index=False)
```

##### Creating function to return Latitude and Longitude by using geocoder

```python
import geocoder
def get_LatitudeLongitude(postal_code):
    coordinate = None
    while(coordinate is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        coordinate = g.latlng
    return coordinate
```

#### Calling Latitude and Longitude for all postal codes in Dataframe

```python
postal_codes = new_df['Postalcode']    
coords = [ get_LatitudeLongitude(postal_code) for postal_code in postal_codes.tolist() ]
coords
```

```python
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
new_df['Latitude'] = df_coords['Latitude']
new_df['Longitude'] = df_coords['Longitude']
new_df.drop(0,axis = 0, inplace=True)
```

```python
def getVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)
            
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(name, lat, lng, v['venue']['name'], v['venue']['location']['lat'], v['venue']['location']['lng'], v['venue']['categories'][0]['name']) for v in results])

    venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    venues.columns = ['Neighbourhood', 'Neighbourhood Latitude', 'Neighbourhood Longitude', 'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category']
    
    return(venues)
```

### Analyzing the Neighbourhood

```python
toronto_onehot = pd.get_dummies(venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighbourhood'] = venues['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot.head()
```

```python
grouped = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
grouped
```

```python
### Printing top 5 venues in each neighbourhood
for neighbour in grouped['Neighbourhood']:
    print("Neighbourhood:",neighbour)
    temp = grouped[grouped['Neighbourhood'] == neighbour].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(5))
    print('\n')
```

### Sorting based on venues and placing in Data frame

```python

def most_common_venues(df, num):
    df_categories = df.iloc[1:]
    df_categories_sorted = df_categories.sort_values(ascending=False)
    return df_categories_sorted.index.values[0:num]
```

```python

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(10):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighbourhoods_venues_sorted = pd.DataFrame(columns=columns)
neighbourhoods_venues_sorted['Neighbourhood'] = grouped['Neighbourhood']

for ind in np.arange(grouped.shape[0]):
    neighbourhoods_venues_sorted.iloc[ind, 1:] = most_common_venues(grouped.iloc[ind, :], 10)

neighbourhoods_venues_sorted.head()
```

### Clustering Neighbours

```python
clustering = grouped.drop('Neighbourhood', 1)

# creating k-means clusters
kmeans = KMeans(n_clusters=5, random_state=0).fit(clustering)

neighbourhoods_venues_sorted.insert(0, 'Cluster_Labels', kmeans.labels_)

toronto_merged = new_df

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighbourhoods_venues_sorted.set_index('Neighbourhood'), on='Neighborhood')

toronto_merged.head()
```

```python
toronto_merged=toronto_merged.dropna()
toronto_merged['Cluster_Labels'] = toronto_merged.Cluster_Labels.astype(int)
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(5)
ys = [i + x + (i*x)**2 for i in range(5)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster_Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker([lat, lon],radius=5,popup=label,color=rainbow[cluster-1],fill=True,fill_color=rainbow[cluster-1],fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
